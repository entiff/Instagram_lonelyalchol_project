```python
# 연관규칙의 결과 보여주는 함수 
def show(of_rule):
    re = list(of_rule)
    rule_info = []
    for i in range(len(re)):
        rule_info = rule_info + [{'frozenset: ': re[i][0]}, {'support: ': re[i][1]}, {'confidence: ': 
                       re[i][2][0][2]} ,{'lift: ': re[i][2][0][3]}, '=========================']
        
    return rule_info
```
